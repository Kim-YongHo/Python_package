## 파일관련 편리한 모듈


- pickle

```python
from glob import glob 

glob('*.ipynb') # 현재 디렉터리에 특정 확장자 파일 목록 반환
glob('*.txt') # 현재 디렉터리에 txt 확장자 파일 목록 반환
glob('../data/*.txt')
glob('../data/*.csv') # 현제 디렉터리 상위 디렉터리의 data 폴더에 있는 csv 파일 과 경로 반환
```

```python
# 해당 파일 리스트가 디렉터인지 확인
from glob import glob
from os.path import isdir

for x in glob('../data/*'): # *은 모든 것을 의미
        if isdir(x):
            print(x, '<DIR> 경로 파일')
        else:
            print(x)
              
```

```python
## data 폴더에 있는 모든 csv 파일을 읽어오기

from glob import glob

# csv 목록 불러오기
file_names = glob("../data/store/*.csv") # 경로 정보까지 가져와서 쉽게 사용할 수 있음(경로명을 포함)
file_names
```

```python
# csv 읽어와서 기존 df에 병합 : concat 함수 사용

total = pd.DataFrame() # 빈 데이터프레임 생성

# 모든 csv 병합하기
for file_name in file_names:
    temp = pd.read_csv(file_name, sep=',', encoding='utf-8')
    total = pd.concat([total, temp])
```

```python
total.shape
```

```python
total.head()
```

```python
total.info()
```

```python
#index 재설정

total.reset_index(inplace=True, drop=True)
```

```python
total.info()
```

```python
# 사용할 컬럼만 선택 : 기본 인덱싱 사용 => df[[컬럼1, 컬럼2,....]]

data = total[['상호명', '지점명', '상권업종대분류명', '상권업종중분류명', '시도명',
             '시군구명', '행정동명']]

data
```
